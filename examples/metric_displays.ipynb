# Metric Displays:

```python
import avt
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from matplotlib import animation
```

## Reliability Display

Fake data:

```python
X, y = make_classification(
    n_features=12, n_informative=10, random_state=1, n_classes=10
)
```

```python
plt.scatter(X[:, 0], X[:, 1], marker="o", c=y, s=25, edgecolor="k")
```

Predicting with LR:

```python
clf = LogisticRegression(max_iter=1000)
clf.fit(X, y)
```

Plotting the Reliability Diagram from the predictions:

```python
y_prob = clf.predict_proba(X)

ax = avt.ReliabilityDisplay.from_predictions(
    y, 
    y_prob,
)

# saving graph
ax[0].figure.savefig('./figures/ReliabilityDisplay_from_predictions.png', bbox_inches='tight', dpi=300)
ax[0].figure.savefig(
    '../docs/source/figures/ReliabilityDisplay_from_predictions.png', bbox_inches='tight', dpi=300
)
```

Or from the estimator directly:

```python
ax = avt.ReliabilityDisplay.from_estimator(
    clf,
    X,
    y,
)

# saving graph
ax[0].figure.savefig('./figures/ReliabilityDisplay_from_estimator.png', bbox_inches='tight', dpi=300)
ax[0].figure.savefig(
    '../docs/source/figures/ReliabilityDisplay_from_estimator.png', bbox_inches='tight', dpi=300
)
```
